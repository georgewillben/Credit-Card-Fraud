## Show the distributions of the features

Now I will create histograms and kernel density estimation plots to show the distribution of each feature.

```python
# Plot histograms and KDEs for each feature
for col in df.drop("Class", axis=1).columns:
    plt.figure(figsize=(12,3))
    plt.ylabel('Probability Density')
    sns.distplot(df[col])
    plt.show()
```

## Show distributions of the features when split by class

Now I will show the distribution of the features when split by class. I will use histograms and kernel density estimation plots again. I will plot vertical lines for the maximum and minimum values for each set to show where they lie.

```python
# Split each feature by class and then plot histograms and KDEs
for col in df.drop("Class", axis=1).columns:
    x0 = df[df.Class == 0][col]
    x1 = df[df.Class == 1][col]
    plt.figure(figsize=(12,3))
    plt.ylabel('Probability Density')
    sns.distplot(x0, bins=50, label='normal')
    sns.distplot(x1, bins=50, label='fraudulent')
    
    # Plot vertical lines
    plt.axvline(x0.min(), label='normal min', linestyle=':')
    plt.axvline(x0.max(), label='normal max', linestyle='--')
    plt.axvline(x1.min(), label='fraud min', color='C1', linestyle=':')
    plt.axvline(x1.max(), label='fraud max', color='C1', linestyle='--')
    
    
    plt.legend()
    plt.show()
```

There are two things to notice here:
* There are times of day when the ratio between normal and fraudlent activity varies largely.
* Principal component features 1, 5, 6, 8, 13, 15, 20, 21, 22, 23, 24, 25, and 26 do not seem to vary much when split by classes.

## Question: Do fraudulent transactions tend to have a higher or lower amount?
I will answer this question by looking at the distribution of the amount column in two slices of the dataframe split by Class.

# Matplotlib & Seaborn: Plot distributions

```python
# show distribution of the Amount column when split by Class

# create slices for non-fraud and fraud transactions
non_fraud = df[df.Class == 0]
fraud = df[df.Class == 1]

# display measures of central tendancy for both slices' Amount column
amount_describe = pd.concat([non_fraud.Amount.describe(), fraud.Amount.describe()], axis=1)
amount_describe.columns=["Non-fraud", "Fruad"]
print(amount_describe)

# create two subplots
f, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(16, 8))

# add title
f.suptitle("Distribution on Amount split by Class")

# plot non-fraud
sns.distplot(non_fraud.Amount, ax=ax1)
ax1.set_title("Non-fraud")

# plot fraud
sns.distplot(fraud.Amount, ax=ax2)
ax2.set_title("Fraud")

plt.show()
```

## Answer: It appears the amount is distributed the same between non-fraudulent and fraudulent transactions.

## Question: Are duplicate rows likely to be fraud?
I will investigate this by seperating the duplicate rows from the rest of the data, then removing the duplicate rows, and finally checking the normalized value counts of the class column for the duplicate and non duplicate data.

```python
def duplicate_fraud(df, title):
    # seperate duplicate and non-duplicate data, drop duplicates and retrieve normalized value counts of the class columns
    duplicates = df[df.duplicated()].Class.value_counts(normalize=True)[1]*100
    non_duplicates = df[df.duplicated() == False].Class.value_counts(normalize=True)[1]*100
 
    # display normalized value counts of the class columns
    cols = ["Duplicated", "Non-duplicated"]
    y = [duplicates, non_duplicates]
    plt.figure(figsize=(16, 4))
    plt.barh(cols ,y)
    plt.title(title)
    for n, vc in enumerate(y):
        plt.text(vc-0.1, n, str(round(vc,3))+"%", color="w")
    plt.show()
```

```python
# investigate whether duplicate transactions have an impact on fraud
duplicate_fraud(df, "Fraud rate in duplicated transactions (percent)")
```

## Answer: Duplicate transactions are about 10 times more likely to be fraud

## Question: Are there transactions that are identical but have a different time value? If so are these transactions likely to be fraud?
This question will be solved the way the previous one was but I will first drop the time column.

```python
time_removed = df.drop("Time", axis=1)
```

```python
# investigate whether duplicate transactions (not including time) have an impact on fraud
duplicate_fraud(time_removed, "Fraud rate in duplicate transactions (ignoring time)")
```

## Answer: It appears that identical transactions with different times do not indicate fraud.

## Question: Is there a time of day where fraudulent transactions are more likely to occur?
For this question I will set up time slots by the hour and see if certain hours have a higher ratio of frauds. This data set does not give me information on the hour of the day so I will only be able to potentially spot a time window where fraud is more prevelent and I won't be able to say what hours of the day that is happening.

```python
# investigate if there is a trend with a specific hour and the ratio of fraud to non fraud

# create bins for cutting the data by the hour
bins = range(0, int(df.Time.max()) + 3600, 3600)

# cut the data
cut_time = pd.cut(df.Time, bins=bins, labels=range(1, len(bins)), include_lowest=True)
cut_time.name = "cut_Time"

# concatanate data
hour_df = pd.concat([df, cut_time], axis=1)

# plot
plt.figure(figsize=(14, 24))
hours = cut_time.unique()

# take normalized value counts for each hour
y = [1 - hour_df[hour_df.cut_Time == h].Class.value_counts(normalize=True)[0] for h in hours] 

plt.barh(hours, y)
plt.axhline(24, label="New Day", color="green")
plt.title("Fraudulent activity ratio per hour")
plt.yticks(hours)
plt.legend()
plt.show()
```

## Answer: There seems to be more fraudulent activity in the earlier part of the 24 periods.
